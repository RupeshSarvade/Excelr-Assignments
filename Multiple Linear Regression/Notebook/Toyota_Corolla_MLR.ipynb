# Building Model

### By using train test split

```python
X = data.iloc[:,1:]
y = data.iloc[:,:1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 40)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test,y_pred)
print(f"mse: {mse},\nrmse: {rmse},\nr2_score: {r2}")
```

## Calculating Vif ( Variance Inflation Factor)

```python
rsq_age = smf.ols("age~KM+HP+cc+Doors+Gears+qt+Weight", data = data).fit().rsquared
vif_age = 1/(1-rsq_age)

rsq_KM = smf.ols("KM~age+HP+cc+Doors+Gears+qt+Weight", data = data).fit().rsquared
vif_KM = 1/(1-rsq_KM)

rsq_HP = smf.ols("HP~KM+age+cc+Doors+Gears+qt+Weight", data = data).fit().rsquared
vif_HP = 1/(1-rsq_HP)

rsq_cc = smf.ols("cc~KM+HP+age+Doors+Gears+qt+Weight", data = data).fit().rsquared
vif_cc = 1/(1-rsq_cc)

rsq_Doors = smf.ols("Doors~KM+HP+cc+age+Gears+qt+Weight", data = data).fit().rsquared
vif_Doors = 1/(1-rsq_Doors)

rsq_Gears = smf.ols("Gears~KM+HP+cc+Doors+age+qt+Weight", data = data).fit().rsquared
vif_Gears = 1/(1-rsq_Gears)

rsq_qt = smf.ols("qt~KM+HP+cc+Doors+Gears+age+Weight", data = data).fit().rsquared
vif_qt = 1/(1-rsq_qt)

rsq_Weight = smf.ols("Weight~KM+HP+cc+Doors+Gears+qt+age", data = data).fit().rsquared
vif_Weight = 1/(1-rsq_Weight)

print(f"vif_age: {vif_age},\nvif_KM: {vif_KM},\nvif_HP: {vif_HP},\nvif_cc: {vif_cc},\nvif_Doors:{vif_Doors},\nvif_Gears: {vif_Gears},\nvif_qt: {vif_qt},\nvif_Weight: {vif_Weight}")
```

#### creating dataframe of vif scores for better understanding

```python
data.columns
```

```python
df = {"variable":['age', 'KM', 'HP', 'cc', 'Doors', 'Gears', 'qt', 'Weight'],"vif":[vif_age,vif_KM,vif_HP,vif_cc,vif_Doors,vif_Gears,vif_qt,vif_Weight]}
vif = pd.DataFrame(df)
vif
```

VIF = 1 (Not correlated) 1 < VIF < 5 (Moderately correlated) VIF >=5 (Highly correlated)

Here, we can consider this features for model generation because,as per the vif scores featires are not highly correlated to each other

## Residual Analysis

### Test for normality of residuals (Q-Q plot / quantile-quantile plot)

```python
#calculating residual value for ecah data point
model.resid
```

```python
import statsmodels.api as sm
qqplot = sm.qqplot(model.resid, line = "q")
plt.title("Normality qqplot of residuals")
plt.show()
```

## Residual Plot for Homoscedasticity

```python
def get_standardized_values(vals):
    return (vals - vals.mean())/vals.std()
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),
            get_standardized_values(model.resid))
plt.title("Residual Plot")
plt.xlabel("Fitted Value")
plt.ylabel("Residual Values")
plt.show()
```

## Residuals vs Regressors

```python
#plooting all variables by using for loop
for i in data.columns:
    try:       
        fig = plt.figure(figsize = (15,7))
        sm.graphics.plot_regress_exog(model,i,fig = fig)
        plt.title(i)
        plt.show()
    except ValueError:
        pass
    else:
        fig = plt.figure(figsize = (15,7))
        sm.graphics.plot_regress_exog(model,i,fig = fig)
        plt.title(i)
        plt.show()
        
```

### Detecting Influencers/Outliers

### Cook's distance

```python
model_influence = model.get_influence()
(c,_) = model_influence.cooks_distance
```

```python
#plotting influencers using stem plot
fig = plt.figure(figsize = (20,7))
plt.stem(np.arange(len(data)),np.round(c,3))
plt.xlabel("Row Index")
plt.ylabel("cook's distance")
plt.show()
```

```python
#cook's distnace shows one influencer point at index 80
(np.argmax(c),np.max(c))
```
